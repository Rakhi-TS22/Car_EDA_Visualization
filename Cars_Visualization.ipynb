```python
fig = plt.subplots(2,2, figsize=(20,10))
plt.subplot(2,2,1)
sns.scatterplot(x=dt['mpg'], y=dt['hp'],data = dt, marker = 'o', hue=dt['brand'])
plt.subplot(2,2,2)
sns.scatterplot(x=dt['mpg'], y=dt['weightlbs'],data = dt, marker = 's',hue=dt['brand'])
plt.subplot(2,2,3)
sns.scatterplot(x=dt['hp'], y=dt['mpg'],data = dt, marker = 's',hue=dt['brand'])
plt.subplot(2,2,4)
sns.scatterplot(x=dt['hp'], y=dt['time-to-60'],data = dt, marker = 's',hue=dt['brand'])
```

```python
plt.subplots(2,2, figsize=(20,10))
plt.subplot(2,2,1)
sns.barplot(x=dt['cylinders'],y=dt['weightlbs'],data = dt, hue = dt['brand'])
plt.subplot(2,2,2)
sns.barplot(x=dt['cylinders'],y=dt['mpg'],data = dt, hue = dt['brand'])
plt.subplot(2,2,3)
sns.barplot(x=dt['cylinders'], y = dt['weightlbs'], data = dt, hue = dt['brand'] )
plt.subplot(2,2,4)
sns.barplot(x= dt['cylinders'],y=dt['time-to-60'], data = dt, hue = dt['brand'])
```

```python
plt.subplots(2,2, figsize=(20,10))
plt.subplot(2,2,1)
sns.barplot(x=dt['year'],y=dt['hp'],data = dt, hue = dt['brand'])
plt.subplot(2,2,2)
sns.barplot(x=dt['year'],y=dt['mpg'],data = dt, hue = dt['brand'])
plt.subplot(2,2,3)
sns.barplot(x=dt['year'], y = dt['weightlbs'], data = dt, hue = dt['brand'] )
plt.subplot(2,2,4)
sns.barplot(x= dt['year'],y=dt['time-to-60'], data = dt, hue = dt['brand'])
```

```python
plt.subplots(2,2, figsize=(20,10))
plt.subplot(2,2,1)
sns.boxplot(x=dt['cylinders'],y=dt['hp'],data = dt, hue = dt['brand'])
plt.subplot(2,2,2)
sns.boxplot(x=dt['cylinders'],y=dt['mpg'],data = dt, hue = dt['brand'])
plt.subplot(2,2,3)
sns.boxplot(x=dt['cylinders'], y = dt['weightlbs'], data = dt, hue = dt['brand'] )
plt.subplot(2,2,4)
sns.boxplot(x= dt['cylinders'],y=dt['time-to-60'], data = dt, hue = dt['brand'])
```
